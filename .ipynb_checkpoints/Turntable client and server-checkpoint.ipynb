```python
##Client for arduino control over network using pygame
import socket
import sys
import pygame
import time

HOST, PORT = "192.168.0.116", 9997

pygame.init()
pygame.event.set_grab(True)

sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
try:
    sock.connect((HOST, PORT))
except:
    raise
    
done = False
while not done:
    data = ""
    events = pygame.event.get()
    for event in events:
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                data = "," 
            if event.key == pygame.K_RIGHT:
                data = "."
            if event.key == pygame.K_SPACE:
                data = " "
            if event.key == pygame.K_ESCAPE:
                done = True
    # Create a socket (SOCK_STREAM means a TCP socket)
    try:
        # Send data
        #print data
        if data != "":
            sock.sendall(data + "\n")
        #print data.strip()
        #Receive data from the server
        #received = sock.recv(1024)
    except:
        raise
    #print "Sent:     {}".format(data)
    #print "Received: {}".format(received)
    time.sleep(1)
sock.close()
print "Client closed"
```

```python
##Server for communicating with arduino
import SocketServer
import socket
import serial
import time

class MyTCPHandler(SocketServer.BaseRequestHandler):
    """
    The RequestHandler class for our server.
    """

    def handle(self):
        # self.request is the TCP socket connected to the client
        self.data = self.request.recv(1024)
        #print "{} wrote:".format(self.client_address[0])
        # just send back the same data, but upper-cased
        while self.data != '':
            ser.write(self.data)
            self.data = self.request.recv(1024)
        #print self.data

HOST, PORT = "10.162.177.215", 9999 #192.168.0.116

#Reset the socket to allow a reconnection
s = socket.socket()
s.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)

# Create the server, binding to host on port 
server = SocketServer.TCPServer((HOST, PORT), MyTCPHandler)

# open usb serial connection to arduino
ser = serial.Serial('/dev/tty.usbmodem1411', 9600)
print "USB port: ", ser.portstr

# Activate the server; this will keep running until you
# interrupt the program with Ctrl-C
server.serve_forever()
        
# close the connection
ser.close()
print "Sequence complete."
```
